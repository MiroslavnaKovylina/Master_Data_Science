#### Filas con porcentaje de valores nulos (f1)

```python
def rows_with_all_nulls_above_threshold(df, threshold_percentage):
    """
    Calcula el número de filas en un DataFrame donde el porcentaje de columnas nulas
    excede un umbral especificado. Si existen filas que cumplen esta condición,
    también las imprime para revisión.

    Parámetros:
    ----------
    df (pd.DataFrame): El DataFrame sobre el que se realizará el análisis.
    threshold_percentage (float): Porcentaje mínimo de columnas nulas en una fila 
                                   para ser considerado.

    Devuelve:
    -------
    int: El número de filas donde el porcentaje de columnas nulas excede el umbral.
    """
    
    # Total de columnas en el DataFrame
    total_columns = df.shape[1]
    
    # Calcular el porcentaje de columnas nulas para cada fila
    null_percentage_per_row = df.isna().sum(axis=1) / total_columns * 100
    
    # Filtrar filas donde el porcentaje de columnas nulas supera el umbral
    all_nulls_df = df[null_percentage_per_row > threshold_percentage]
    
    # Contar el número de filas que cumplen la condición
    num_above_threshold = all_nulls_df.shape[0]
    
    # Imprimir el resumen y las filas que cumplen la condición, si las hay
    if num_above_threshold > 0:
        print(f"Hay {num_above_threshold} filas con más del {threshold_percentage}% de columnas nulas.")
        print("\nLas filas correspondientes son:")
        print(all_nulls_df)
    else:
        print(f"No hay filas con más del {threshold_percentage}% de columnas nulas.")
    
    return num_above_threshold
```

#### Reducir memoria (f2)

```python
def reducir_memoria(df, nan_in_df=True):
    """
    Optimiza el uso de memoria de un DataFrame reduciendo el tipo de datos de las columnas,
    siempre que sea posible. Convierte columnas numéricas a tipos más pequeños y columnas
    categóricas si cumplen ciertos criterios.

    Parámetros:
    ----------
    df : pd.DataFrame
        El DataFrame que se optimizará.
    nan_in_df : bool, opcional
        Indica si hay valores NaN en el DataFrame. Si es True, las columnas con valores NaN
        se convierten a tipos compatibles como 'Int64'. Por defecto es True.

    Retorna:
    -------
    pd.DataFrame
        El DataFrame optimizado con menor uso de memoria.
    """
    # Calcular el uso inicial de memoria
    inicio_memoria = df.memory_usage().sum() / 1024**2
    print(f"Memoria inicial del DataFrame: {inicio_memoria:.2f} MB")

    # Iterar a través de las columnas del DataFrame
    for col in df.columns:
        col_tipo = df[col].dtype

        # Si la columna es de tipo numérico
        if pd.api.types.is_numeric_dtype(col_tipo):
            if pd.api.types.is_integer_dtype(col_tipo):
                # Reducir enteros al tipo más pequeño que pueda almacenar sus valores
                if df[col].min() >= np.iinfo(np.int16).min and df[col].max() <= np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif df[col].min() >= np.iinfo(np.int32).min and df[col].max() <= np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
            elif pd.api.types.is_float_dtype(col_tipo):
                # Reducir flotantes a float32
                df[col] = df[col].astype(np.float32)
                # Convertir a Int64 si tiene exactamente dos valores únicos (compatible con NaN)
                if df[col].nunique() == 2:
                    df[col] = df[col].astype('Int64')
                # Convertir a enteros más pequeños si no hay NaN y tiene pocos valores únicos
                if (not nan_in_df) and (df[col].nunique() < 4):
                    df[col] = df[col].astype('int16')

        # Si la columna es de tipo datetime, no se realiza conversión
        elif pd.api.types.is_datetime64_any_dtype(col_tipo):
            continue

        # Si la columna es de tipo objeto
        elif pd.api.types.is_object_dtype(col_tipo):
            num_unicos = df[col].nunique()
            num_total = len(df[col])
            # Convertir a categoría si los valores únicos son menos del 50% del total
            if num_unicos / num_total < 0.5:
                df[col] = df[col].astype('category')

    # Calcular el uso final de memoria
    fin_memoria = df.memory_usage().sum() / 1024**2
    print(f"Memoria final del DataFrame: {fin_memoria:.2f} MB")
    print(f"Reducción del {100 * (inicio_memoria - fin_memoria) / inicio_memoria:.2f}%")

    return df
```

#### Identificar grupos con múltiples valores únicos en una columna específica (f3)

```python
def count_groups_with_multiple_unique_values_and_nan(df, groupby_column, value_column, print_groups=True):
    """
    Identifica grupos con más de un valor único en una columna específica, 
    considerando tanto valores no nulos como aquellos que incluyen al menos un NaN. 
    Muestra los resultados si se solicita.

    Parámetros:
    - df (DataFrame): El DataFrame de entrada.
    - groupby_column (str): La columna utilizada para agrupar los datos.
    - value_column (str): La columna cuyos valores únicos se analizarán dentro de cada grupo.
    - print_groups (bool): Si es True, imprime los grupos que cumplen las condiciones 
                           (por defecto es True).

    Retorna:
    - count (int): Número de grupos con más de un valor único en `value_column` 
                   que incluye al menos un NaN.
    - groups_with_more_than_one_and_nan (Series): Grupos con más de un valor único 
                                                  y al menos un NaN.
    - groups_with_more_than_one (Series): Grupos con más de un valor único, 
                                          independientemente de si contienen NaN.

    Pasos del análisis:
    1. Calcula el número de valores únicos, incluyendo NaN, en `value_column` 
       para cada grupo definido por `groupby_column`.
    2. Filtra los grupos que tienen más de un valor único.
    3. Identifica los grupos que contienen al menos un NaN en `value_column`.
    4. Filtra los grupos con más de un valor único y que además incluyen un NaN.
    5. Muestra estadísticas y resultados de los grupos que cumplen las condiciones.
    """
    # Calcular el número de valores únicos en la columna especificada para cada grupo, incluyendo NaN
    nunique_with_nan = df.groupby(groupby_column)[value_column].nunique(dropna=False)
    
    # Filtrar los grupos que tienen más de un valor único
    groups_with_more_than_one = nunique_with_nan[nunique_with_nan > 1]
    
    # Identificar los grupos que contienen al menos un valor NaN
    groups_with_nan = df[df[value_column].isna()].groupby(groupby_column).size().index
    
    # Filtrar grupos que tienen más de un valor único y al menos un NaN
    groups_with_more_than_one_and_nan = groups_with_more_than_one.loc[groups_with_more_than_one.index.intersection(groups_with_nan)]

    # Contar cuántos grupos cumplen ambas condiciones
    count = groups_with_more_than_one_and_nan.count()

    # Imprimir resultados
    print(f"Número de clientes con más de un valor único en '{value_column}': {len(groups_with_more_than_one)}")
    print(f"Número de clientes con más de un valor único y al menos un NaN: {count}")
    if len(groups_with_more_than_one_and_nan) > 0 and print_groups:
        print("Clientes con más de un valor único y al menos un NaN:")
        print(groups_with_more_than_one_and_nan)

    return count, groups_with_more_than_one_and_nan, groups_with_more_than_one
```

#### Visualizar distribución (f4)

```python
def visualizar_distribucion(df, columna, eje='y', escala_logaritmica=False, figsize=(6, 3), rotar_ticks=False, grados_rotacion=0, xlabel=None, ylabel=None, order=None):
    """
    Genera un gráfico de distribución para visualizar la frecuencia de los valores en una columna específica.

    Parámetros:
    - df (DataFrame): El DataFrame que contiene los datos a analizar.
    - columna (str): Nombre de la columna cuya distribución será visualizada.
    - eje (str): Define el eje donde se mostrará la distribución ('x' o 'y'). Por defecto, 'y'.
    - escala_logaritmica (bool): Si es True, aplica escala logarítmica al eje numérico. Por defecto, False.
    - figsize (tuple): Dimensiones de la figura (ancho, alto) en pulgadas. Por defecto, (6, 3).
    - rotar_ticks (bool): Si es True, rota las etiquetas de los ticks en el eje seleccionado. Por defecto, False.
    - grados_rotacion (int): Ángulo de rotación para los ticks si `rotar_ticks` es True. Por defecto, 0.
    - xlabel (str): Etiqueta personalizada para el eje X. Si es None, se asigna automáticamente.
    - ylabel (str): Etiqueta personalizada para el eje Y. Si es None, se asigna automáticamente.
    - order (list): Orden específico para las categorías en el gráfico. Por defecto, None.

    Retorna:
    - Gráfico de barras que muestra la distribución de la columna seleccionada.

    Notas:
    - Si `escala_logarítmica` es True, se aplicará una escala logarítmica al eje numérico (X o Y dependiendo del gráfico).
    - Si el parámetro `rotar_ticks` es True, las etiquetas del eje serán rotadas según `grados_rotacion`.
    - Este método utiliza `sns.countplot` para la visualización.
    """
    # Configurar el tamaño de la figura
    plt.figure(figsize=figsize)

    # Crear el gráfico de distribución basado en el eje especificado
    if eje == 'x':
        sns.countplot(x=columna, data=df, order=order)
    elif eje == 'y':
        sns.countplot(y=columna, data=df, order=order)
    else:
        # Validación de parámetros: el eje debe ser 'x' o 'y'
        raise ValueError("El parámetro 'eje' debe ser 'x' o 'y'.")

    # Etiqueta base para el eje numérico
    axis_name = "Frecuencia"

    # Aplicar escala logarítmica si se especifica
    if escala_logaritmica:
        axis_name += " (escala logarítmica)"
        if eje == 'x':
            # Aplicar escala logarítmica al eje Y
            plt.yscale('log')
        else:
            # Aplicar escala logarítmica al eje X
            plt.xscale('log')

    # Rotar las etiquetas de los ticks si se solicita
    if rotar_ticks:
        if eje == 'x':
            # Rotar los ticks del eje X
            plt.xticks(rotation=grados_rotacion)
        else:
            # Rotar los ticks del eje Y
            plt.yticks(rotation=grados_rotacion)

    # Configurar títulos y etiquetas del gráfico
    plt.title(f'Distribución de {columna}')
    plt.xlabel(xlabel if xlabel else (axis_name if eje == 'y' else columna.capitalize()))
    plt.ylabel(ylabel if ylabel else (columna.capitalize() if eje == 'y' else axis_name))

    # Ajustar el diseño para evitar superposición de elementos
    plt.tight_layout()

    # Mostrar el gráfico
    plt.show()
```

#### Visualizar histograma (f5)

```python
def visualizar_histograma(df, columna, bins=15, escala_logaritmica=False, figsize=(6, 3), xlabel=None, ylabel=None):
    """
    Visualiza la distribución de una columna numérica en un DataFrame mediante un histograma.

    Parámetros:
    - df (DataFrame): El DataFrame que contiene los datos a analizar.
    - columna (str): El nombre de la columna numérica cuya distribución será visualizada.
    - bins (int): Número de intervalos (bins) en el histograma. Por defecto, 15.
    - escala_logaritmica (bool): Si es True, aplica escala logarítmica al eje Y para resaltar datos en rangos pequeños. Por defecto, False.
    - figsize (tuple): Dimensiones de la figura (ancho, alto) en pulgadas. Por defecto, (6, 3).
    - xlabel (str): Etiqueta personalizada para el eje X. Si no se proporciona, se asignará automáticamente con el nombre de la columna.
    - ylabel (str): Etiqueta personalizada para el eje Y. Si no se proporciona, se asignará como "Frecuencia" o con una indicación de escala logarítmica.

    Retorna:
    - Un gráfico de histograma que representa la distribución de valores en la columna especificada.

    Notas:
    - Si `escala_logarítmica` es True, el eje Y se muestra en escala logarítmica para enfatizar las diferencias en valores con baja frecuencia.
    - Utiliza `sns.histplot` para generar el histograma, que proporciona una visualización clara de las distribuciones.
    - Las etiquetas del eje X e Y pueden personalizarse a través de los parámetros `xlabel` y `ylabel`.
    """
    # Configurar el tamaño de la figura
    plt.figure(figsize=figsize)
    
    # Crear el histograma
    sns.histplot(x=df[columna], bins=bins)

    # Etiqueta base para el eje Y
    axis_name = "Frecuencia"
    
    # Aplicar escala logarítmica al eje Y si se especifica
    if escala_logaritmica:
        axis_name += " (escala logarítmica)"
        plt.yscale('log')
    
    # Configurar título y etiquetas
    plt.title(f'Distribución de {columna}')
    plt.xlabel(xlabel if xlabel else columna.capitalize())
    plt.ylabel(ylabel if ylabel else axis_name)
    
    # Ajustar el diseño para evitar superposición de elementos
    plt.tight_layout()
    
    # Mostrar el gráfico
    plt.show()
```

#### Calcular frecuencia acumulada (f6)

```python
def calcular_frecuencia_acumulada(df, columna, digit_to_round=4):
    """
    Calcula el porcentaje de representación de cada valor en una columna específica, 
    junto con la frecuencia acumulada, y redondea los resultados.

    Parámetros:
    - df (pd.DataFrame): El DataFrame que contiene los datos a analizar.
    - columna (str): El nombre de la columna cuyos valores se analizarán.
    - digit_to_round (int): El número de decimales para redondear las frecuencias. Por defecto es 4.

    Retorna:
    - pd.DataFrame: Un DataFrame con las siguientes columnas:
        - `columna`: Valores únicos de la columna original.
        - `Frecuencia(%)`: Porcentaje que representa cada valor único en la columna.
        - `Frecuencia_acumulada(%)`: Porcentaje acumulado de las frecuencias.

    Notas:
    - La frecuencia relativa se calcula dividiendo la cuenta de cada valor único entre el total.
    - Los valores se multiplican por 100 para representarlos en porcentaje.
    - El cálculo de la frecuencia acumulada permite identificar rápidamente cuántos valores contribuyen a una fracción específica del total.
    """
    # Calcular la frecuencia relativa (normalizada)
    df_range = df[columna].value_counts(normalize=True).reset_index(name='Frecuencia')
    
    # Calcular la frecuencia acumulada
    df_range['Frecuencia_acumulada'] = df_range['Frecuencia'].cumsum()
    
    # Convertir las frecuencias a porcentaje y redondear
    df_range['Frecuencia'] = df_range['Frecuencia'].mul(100).round(digit_to_round)
    df_range['Frecuencia_acumulada'] = df_range['Frecuencia_acumulada'].mul(100).round(digit_to_round)
    
    # Renombrar las columnas
    df_range.columns = [columna, 'Frecuencia(%)', 'Frecuencia_acumulada(%)']
    
    return df_range
```

#### Categorizar 'age' (f7)

```python
def categorizar_age(age):
    """
    Categoriza un valor de edad en rangos definidos.

    Parámetros:
    - age (int o float): La edad que se desea categorizar.

    Retorna:
    - str: Una cadena que representa el rango de edad al que pertenece el valor de entrada.

    Categorías:
    - '0-20': Para edades entre 0 y 20 años (inclusive).
    - '21-35': Para edades entre 21 y 35 años (inclusive).
    - '36-50': Para edades entre 36 y 50 años (inclusive).
    - '51-65': Para edades entre 51 y 65 años (inclusive).
    - '66+': Para edades mayores a 66 años.
    """
    # Verificar si la edad es menor o igual a 20
    if age <= 20:
        return '0-20'
    # Verificar si la edad está entre 21 y 35
    elif age <= 35:
        return '21-35'
    # Verificar si la edad está entre 36 y 50
    elif age <= 50:
        return '36-50'
    # Verificar si la edad está entre 51 y 65
    elif age <= 65:
        return '51-65'
    # Si la edad es mayor a 66
    else:
        return '66+'
```

```python
def categorizar_salario(salario):
    """
    Categoriza un valor de salario en rangos definidos.

    Parámetros:
    - salario (float o int): El salario que se desea categorizar.

    Retorna:
    - str: Una cadena que representa el rango de salario al que pertenece el valor de entrada.

    Categorías:
    - 'Desconocido': Para valores de salario iguales a 0.
    - '< 50k': Para salarios menores a 50,000.
    - '50-100k': Para salarios entre 50,000 y 100,000.
    - '100-200k': Para salarios entre 100,000 y 200,000.
    - '> 200k': Para salarios mayores a 200,000.
    """
    # Verificar si el salario es igual a 0
    if salario == 0:
        return 'Desconocido'
    # Verificar si el salario es menor a 50,000
    elif salario < 50000:
        return '< 50k'
    # Verificar si el salario está entre 50,000 y 100,000
    elif salario < 100000:
        return '50-100k'
    # Verificar si el salario está entre 100,000 y 200,000
    elif salario < 200000:
        return '100-200k'
    # Si el salario es mayor a 200,000
    else:
        return '> 200k'
```

#### Clusterizar y generar estadísticas descriptivas (f9)

```python
def clusterize_and_analyze_mix(df_Task2, k, var_cluster, random_state=3, 
                           scaler_name="Standard", 
                           standard_columns=None, robust_columns=None, minmax_columns=None, binary_columns=None):
    """
    Escala el DataFrame, aplica KMeans y genera un análisis por cluster.

    Args:
        df_Task2 (pd.DataFrame): DataFrame original a procesar.
        k (int): Número de clusters para KMeans.
        var_cluster (list): Variables utilizadas para el clustering.
        random_state (int): Estado aleatorio para reproducibilidad.
        scaler_name (str): Nombre del escalador único 'Standard', 'Robust', 'MinMax' o 'Mixed' para combinar escaladores.
        standard_columns (list): Columnas a escalar con StandardScaler (si scaler_name="Mixed").
        robust_columns (list): Columnas a escalar con RobustScaler (si scaler_name="Mixed").
        minmax_columns (list): Columnas a escalar con MinMaxScaler (si scaler_name="Mixed").
        binary_columns (list): Columnas binarias que no se escalarán.

    Returns:
        pd.DataFrame: DataFrame original con etiquetas de clusters asignadas.
        pd.DataFrame: DataFrame estilizado con estadísticas descriptivas por cluster.
    """

    # Validar parámetros
    if scaler_name not in ["Standard", "Robust", "MinMax", "Mixed"]:
        raise ValueError(f"Escalador '{scaler_name}' no reconocido. Usa 'Standard', 'Robust', 'MinMax' o 'Mixed'.")

    # Crear copia del DataFrame para evitar modificar el original
    df_Task2_copy = df_Task2.copy(deep=True)

    if scaler_name == "Mixed":
        if not (standard_columns or robust_columns or minmax_columns):
            raise ValueError("Debe proporcionar listas para 'standard_columns', 'robust_columns' y 'minmax_columns' cuando scaler_name es 'Mixed'.")
        
        # Escalar con los escaladores correspondientes
        scaled_standard = StandardScaler().fit_transform(df_Task2_copy[standard_columns]) if standard_columns else None
        scaled_robust = RobustScaler().fit_transform(df_Task2_copy[robust_columns]) if robust_columns else None
        scaled_minmax = MinMaxScaler().fit_transform(df_Task2_copy[minmax_columns]) if minmax_columns else None

        # Crear DataFrame de las columnas escaladas
        df_Task2_scaled = pd.DataFrame(index=df_Task2_copy.index)
        
        if standard_columns:
            df_Task2_scaled[standard_columns] = scaled_standard
        if robust_columns:
            df_Task2_scaled[robust_columns] = scaled_robust
        if minmax_columns:
            df_Task2_scaled[minmax_columns] = scaled_minmax
        if binary_columns:
            df_Task2_scaled[binary_columns] = df_Task2_copy[binary_columns]

    else:
        # Escaladores únicos
        scalers = {
            "Standard": StandardScaler(),
            "Robust": RobustScaler(),
            "MinMax": MinMaxScaler()
        }
        scaler = scalers[scaler_name]
        df_Task2_scaled = scaler.fit_transform(df_Task2_copy[var_cluster])

    # Aplicar KMeans
    kmeans = KMeans(n_clusters=k, random_state=random_state)
    clusters = kmeans.fit_predict(df_Task2_scaled)

    # Asignar las etiquetas de cluster al DataFrame original
    df_Task2_copy['cluster'] = clusters

    # Generar una fila con los tamaños de clusters
    tamaño_clusters = df_Task2_copy.groupby("cluster").size().to_frame().T
    tamaño_clusters.index = pd.MultiIndex.from_tuples([('tamaño', '')])
    tamaño_clusters = np.round(tamaño_clusters, 0)

    # Crear DataFrame vacío para estadísticas descriptivas
    segmento_df = pd.DataFrame()

    # Calcular estadísticas descriptivas por cluster
    for col in var_cluster:
        resumen_data = df_Task2_copy[['cluster', col]].groupby("cluster").describe()
        resumen_data = np.round(resumen_data, 2)  # Redondear a 2 decimales
        resumen_data = resumen_data.T.iloc[1:]  # Eliminar count y transponer para mantener el formato
        segmento_df = pd.concat([segmento_df, resumen_data])

    # Asegurar que el índice de 'tamaño_clusters' y 'segmento_df' sean consistentes
    segmento_df.index = pd.MultiIndex.from_tuples(segmento_df.index)

    # Concatenar tamaños de clusters con estadísticas descriptivas
    segmento_df = pd.concat([tamaño_clusters, segmento_df], axis=0)

    # Calcular las métricas
    metrics = {
        "Scaler": [scaler_name],
        "Num Clusters": [k],
        "Inertia (WCSS)": [kmeans.inertia_],
        "CH Score": [calinski_harabasz_score(df_Task2_scaled, clusters)],
        "DBI Score": [davies_bouldin_score(df_Task2_scaled, clusters)],
    }

    # Convertir resultados en DataFrame
    metrics_df = pd.DataFrame(metrics)

    # Imprimir las métricas
    print(f"Métricas:\n{metrics_df.to_string(index=False)}\nVariables utilizadas para el análisis de clusters: {', '.join(var_cluster)}")

    return df_Task2_copy, metrics_df, segmento_df
```

#### Formatear valores de los clusters (f10)

```python
def formato(val):
    """
    Da formato a un valor dependiendo de su tipo:
    - Para números enteros, los devuelve sin decimales.
    - Para números flotantes, los devuelve con 2 decimales.
    - Para valores no numéricos, los devuelve sin cambios.

    Parámetros:
    - val: cualquier tipo de dato
        El valor a formatear.

    Retorno:
    - str o tipo original del valor
        El valor formateado como cadena (para números) o el valor original (para no numéricos).
    """
    # Comprobar si el valor es un número entero o flotante
    if isinstance(val, (int, float)):
        # Si el valor es un número entero, mostrar sin decimales
        if val == int(val):
            return f"{int(val)}"
        # Si el valor es un número flotante, mostrar con 2 decimales
        else:
            return f"{val:.2f}"
    # Si el valor no es numérico, devolver el valor original
    return val
```

#### Función para separar datos en numéricos y categóricas (f11)

```python
def separar_numericas_categoricas(df):
    """
    Separa las columnas de un DataFrame en dos listas: numéricas y categóricas.

    Parámetros:
    - df: pandas.DataFrame
        DataFrame cuyos tipos de columnas se van a analizar.

    Retorno:
    - numerical_columns: list
        Lista de nombres de columnas que contienen datos numéricos.
    - categorical_columns: list
        Lista de nombres de columnas que contienen datos no numéricos.
    """
    # Seleccionar todas las columnas de tipo numérico
    numerical_columns = df.select_dtypes(include="number").columns.to_list()

    # Seleccionar todas las columnas que no son de tipo numérico
    categorical_columns = df.select_dtypes(exclude="number").columns.to_list()

    # Devolver las dos listas: columnas numéricas y categóricas
    return numerical_columns, categorical_columns
```

#### Calcular las correlaciones entre variables categóricas y variables binarias objetivo (f12)

```python
def calculate_categorical_correlations(df, categorical_cols, target_cols):
    """
    Calcula las correlaciones entre variables categóricas y una o varias variables binarias objetivo.

    Parameters:
    - df (pd.DataFrame): DataFrame que contiene las columnas.
    - categorical_cols (list): Lista de nombres de columnas categóricas.
    - target_cols (list): Lista de nombres de columnas binarias objetivo.

    Returns:
    - pd.DataFrame: Tabla de correlaciones.
    """
    def cramers_v(cat_var, target_var):
        """Calcula Cramer's V entre una variable categórica y una binaria."""
        confusion_matrix = pd.crosstab(cat_var, target_var)
        chi2 = chi2_contingency(confusion_matrix)[0]
        n = confusion_matrix.sum().sum()
        return np.sqrt(chi2 / (n * (min(confusion_matrix.shape) - 1)))

    # Crear un DataFrame para almacenar las correlaciones
    correlations = pd.DataFrame(index=categorical_cols, columns=target_cols)

    for target_col in target_cols:
        for cat_col in categorical_cols:
            correlations.loc[cat_col, target_col] = cramers_v(df[cat_col], df[target_col])

    # Convertir a tipo numérico para consistencia
    correlations = correlations.astype(float)

    return correlations
```

#### Lista de columnas numéricas a excluir por ser constantes o Altomente desbalanceadas (f13)

```python
def get_columns_to_exclude(df):
    """
    Analiza un DataFrame y devuelve una lista de columnas a excluir del análisis,
    considerando solo aquellas que tienen 3 o menos valores únicos, y evaluando
    si son constantes o Altomente desbalanceadas.

    Args:
        df (pd.DataFrame): DataFrame a analizar.

    Returns:
        list: Lista de columnas a excluir.
    """
    columns_to_exclude = []
    exclusion_reasons = {}

    # Analizar columnas con 3 o menos valores únicos
    for col in df.columns:
        unique_values = df[col].nunique()

        if unique_values <= 3:
            value_counts = df[col].value_counts()

            if len(value_counts) == 1:
                # Columna constante
                columns_to_exclude.append(col)
                exclusion_reasons[col] = "Columna constante (un único valor)"
            elif value_counts.min() / value_counts.sum() < 0.01:
                # Columna muy desbalanceada (menos del 1%)
                columns_to_exclude.append(col)
                print(value_counts)
                exclusion_reasons[col] = "Columna muy desbalanceada (frecuencia menor al 1%)"

    # Eliminar duplicados en la lista de exclusión
    columns_to_exclude = list(set(columns_to_exclude))

    # Mostrar resultados
    print("Columnas excluidas del análisis y razones:")
    for col in columns_to_exclude:
        print(f"- {col}: {exclusion_reasons.get(col, 'Razón no especificada')}")

    return columns_to_exclude
```

#### Calcular métricas y guardar los resultados en un DataFrame (f14)

```python
def calcular_metricas(modelo, nombre, X_test, y_test, metricas_modelos):
    """
    Calcula varias métricas de evaluación para un modelo y las guarda en un DataFrame.

    Parámetros:
    - modelo: Objeto del modelo entrenado que se evaluará.
    - nombre: Nombre del modelo (str), usado para identificar las métricas en el DataFrame.
    - X_test: Conjunto de datos de características de prueba.
    - y_test: Conjunto de etiquetas reales para comparación.
    - metricas_modelos: DataFrame existente donde se agregarán las métricas del modelo actual.

    Retorno:
    - metricas_modelos: DataFrame actualizado con las métricas del modelo actual.
    """
    # Predecir las etiquetas
    y_pred = modelo.predict(X_test)

    # Calcular las métricas
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    # Para roc_auc_score, hay que predecir probabilidades
    if hasattr(modelo, "predict_proba"):
        y_prob = modelo.predict_proba(X_test)[:, 1]
        roc_auc = roc_auc_score(y_test, y_prob)
    else:
        roc_auc = "N/A"  # Para los modelos sin predict_proba

    # Guardar las métricas en el DataFrame
    nuevo_registro = pd.DataFrame({'modelo': [nombre],
                                   'accuracy': [accuracy],
                                   'precision': [precision],
                                   'recall': [recall],
                                   'f1_score': [f1],
                                   'roc_auc': [roc_auc]})

    metricas_modelos = pd.concat([metricas_modelos, nuevo_registro], ignore_index=True)

    # Ordenar el DataFrame por roc_auc
    metricas_modelos = metricas_modelos.sort_values(by="roc_auc", ascending=False)
    return metricas_modelos
```

#### Función para entrenar y evaluar múltiples modelos (f15)

```python
def torneo_modelos(X_train, y_train, X_test, y_test,print_adjust=False):
    """
    Función para entrenar y evaluar múltiples modelos, comparándolos en base a métricas clave.

    Parámetros:
    - X_train: Conjunto de características para entrenamiento.
    - y_train: Conjunto de etiquetas para entrenamiento.
    - X_test: Conjunto de características para prueba.
    - y_test: Conjunto de etiquetas para prueba.

    Retorno:
    - metricas_producto: DataFrame con las métricas de desempeño de cada modelo.
    """

    # Definir la lista de modelos con sus nombres y configuraciones iniciales
    modelos_producto = [  
        ('RandomForest', RandomForestClassifier()),  # Modelo de Bosque Aleatorio
        ('ExtraTrees', ExtraTreesClassifier()),  # Modelo de Árboles Extra
        ('LogisticRegression Ridge', LogisticRegression(penalty='l2', solver='lbfgs')),  # Regresión Logística con regularización Ridge
        ('LogisticRegression Lasso', LogisticRegression(penalty='l1', solver='saga')),  # Regresión Logística con regularización Lasso
        ('LGBMClassifier', LGBMClassifier(verbose=0)),  # Modelo LightGBM
        ('CatBoostClassifier', CatBoostClassifier(verbose=0)),  # Modelo CatBoost
        ('XGBClassifier', XGBClassifier())  # Modelo XGBoost
    ]

    # Crear un DataFrame vacío para almacenar las métricas de cada modelo
    metricas_producto = pd.DataFrame(columns=['modelo', 'accuracy', 'precision', 'recall', 'f1_score', 'roc_auc'])

    # Iterar sobre la lista de modelos para entrenarlos y calcular sus métricas
    for nombre, modelo_producto in modelos_producto:
        # Entrenar el modelo con el conjunto de entrenamiento
        if print_adjust: print(f"Ajustando modelo: {nombre}")
        modelo_producto.fit(X_train, y_train)

        # Hacer la predicción sobre el conjunto de prueba y calcular las métricas
        metricas_producto = calcular_metricas(modelo_producto, nombre, X_test, y_test, metricas_producto) # f14

    # Devolver el DataFrame con las métricas de desempeño para todos los modelos
    return metricas_producto
```

##### 5.2.8.1 Creación de columnas de diferencia entre meses consecutivos
[Volver al índice](#index)

Para analizar las ventas de productos financieros, utilizamos el método diff de pandas para determinar si, entre dos meses consecutivos, un cliente ha adquirido o dado de baja un producto. Este cambio se refleja en el historial de transacciones como una variación en las columnas correspondientes a productos financieros: de 0 a 1 (adquisición) o de 1 a 0 (baja).

En el caso de los nuevos clientes que ingresaron al sistema durante el mes actual, es posible identificar adquisiciones directas verificando si alguna columna de productos financieros tiene un valor de 1.

Por otro lado, para los clientes cuya información del mes anterior no está disponible y que no son nuevos, eliminamos las filas correspondientes a dicho mes. Esto se debe a que las columnas derivadas de producto_diff contendrán valores NaN. Estos valores no pueden ser imputados de forma precisa, ya que cualquier valor imputado no reflejaría con fidelidad las transacciones reales del cliente, afectando la calidad del análisis posterior.

```python
# Lista de productos financieros
products_list=['short_term_deposit','loans','mortgage','funds','securities','long_term_deposit','credit_card','payroll','pension_plan','payroll_account','emc_account','debit_card','em_account_p','em_acount']
```

```python
# Contar el número de filas por cliente
cliente_con_una_fila = df_all.groupby('pk_cid')['pk_cid'].transform('count') == 1

# Filtrar los clientes que solo tienen una fila
df_clientes_una_fila = df_all[cliente_con_una_fila]

# Imprimir la información
print(f"Número total de clientes con solo un registro: {df_clientes_una_fila['pk_cid'].nunique()}")
print(f"De ellos, clientes con datos del mes de contratación: {df_clientes_una_fila[df_clientes_una_fila['months_from_contract']==0]['pk_cid'].nunique()}")
```

Mantenemos la información de los clientes nuevos y eliminamos a los clientes no nuevos, ya que no se dispone de datos del mes anterior para estos últimos.

```python
# Ordenar por cliente y fecha
df_all = df_all.sort_values(by=['pk_cid', 'pk_partition'])

# Crear las columnas de diferencias consecutivas para todos los productos en products_list e imputar los casos donde 'months_from_contract' == 0
diff_columns = {
    product: df_all.groupby('pk_cid')[product].diff().where(df_all['months_from_contract'] != 0, df_all[product])
    for product in products_list
}

# Agregar las nuevas columnas al DataFrame original
df_all = df_all.assign(**{f"{product}_diff": diff_columns[product] for product in products_list})

# Eliminar las filas con valores NaN en las columnas producto_diff que corresponden a clientes no nuevos y reindexar el DataFrame 
df_all = df_all.dropna().reset_index(drop=True)
```

##### 5.2.8.3 Agrupación de productos financieros por categorías
[Volver al índice](#index)

Agrupamos los productos financieros en las siguientes categorías: cuentas, depósitos, préstamos y créditos, inversiones, y tarjetas. Este enfoque permite analizar las ventas y cancelaciones, proporcionando una visión más clara del comportamiento de los clientes en relación con cada categoría de producto.

```python
# Crear las listas de productos por 5 categorías
cuentas = ['em_account_p_diff','em_acount_diff', 'emc_account_diff', 'payroll_account_diff']
depositos = ['short_term_deposit_diff', 'long_term_deposit_diff']
prestamos_creditos = ['loans_diff', 'mortgage_diff']
inversiones = ['funds_diff', 'securities_diff', 'pension_plan_diff']
tarjetas = ['credit_card_diff', 'debit_card_diff']

# Sumar productos vendidos por cada categoría y convertir a int16
df_all['cuentas_sales'] = df_all[cuentas].where(df_all[cuentas] > 0, 0).sum(axis=1).astype('int16')
df_all['depositos_sales'] = df_all[depositos].where(df_all[depositos] > 0, 0).sum(axis=1).astype('int16')
df_all['prestamos_creditos_sales'] = df_all[prestamos_creditos].where(df_all[prestamos_creditos] > 0, 0).sum(axis=1).astype('int16')
df_all['inversiones_sales'] = df_all[inversiones].where(df_all[inversiones] > 0, 0).sum(axis=1).astype('int16')
df_all['tarjetas_sales'] = df_all[tarjetas].where(df_all[tarjetas] > 0, 0).sum(axis=1).astype('int16')

# Sumar productos dados de baja por cada categoría y convertir a int16
df_all['cuentas_cancelation'] = df_all[cuentas].where(df_all[cuentas] < 0, 0).sum(axis=1).astype('int16')
df_all['depositos_cancelation'] = df_all[depositos].where(df_all[depositos] < 0, 0).sum(axis=1).astype('int16')
df_all['prestamos_creditos_cancelation'] = df_all[prestamos_creditos].where(df_all[prestamos_creditos] < 0, 0).sum(axis=1).astype('int16')
df_all['inversiones_cancelation'] = df_all[inversiones].where(df_all[inversiones] < 0, 0).sum(axis=1).astype('int16')
df_all['tarjetas_cancelation'] = df_all[tarjetas].where(df_all[tarjetas] < 0, 0).sum(axis=1).astype('int16')
```

```python
# Crear 2 nuevas columnas de ventas para las 3 categorías 'ahorro_inversion' y 'financiacion', además de la ya existente 'cuentas_sales'
df_all['ahorro_inversion_sales'] = (df_all['depositos_sales'] + df_all['inversiones_sales'])
df_all['financiacion_sales'] = (df_all['prestamos_creditos_sales'] + df_all['tarjetas_sales'])
```

```python
# Crear columnas que sumen el total de productos vendidos
df_all['products_sales'] = (df_all['cuentas_sales'] + df_all['ahorro_inversion_sales'] + df_all['financiacion_sales'])
# Crear columnas que sumen el total de productos dados de baja
df_all['products_cancelation'] = (df_all['cuentas_cancelation'] + df_all['depositos_cancelation'] + df_all['prestamos_creditos_cancelation'] + df_all['inversiones_cancelation'] + df_all['tarjetas_cancelation'])
```

Al comparar distintos escaladores, se observa que el StandardScaler ofrece los mejores resultados. Su funcionamiento se basa en transformar cada variable para que tenga una media de 0 y una desviación estándar de 1.
Las razones por las cuales este escalador parece ser el más efectivo son:
- Presencia limitada de valores atípicos:  
El análisis de las variables no revela una cantidad significativa de outliers extremos que puedan distorsionar la media o la desviación estándar.
Variables como 'age', 'months_from_contract', 'PIB_2018' y las ventas del último año muestran concentraciones claras en rangos específicos, sin colas extremas significativas.
- Escalas heterogéneas:  
La variable 'PIB_2018' presenta una escala considerablemente más amplia en comparación con otras.
Por otro lado, las variables de ventas ('cuentas_sales_last_year', 'ahorro_inversion_sales_last_year', etc.) y las segmentaciones tienen escalas más reducidas.
Aplicar StandardScaler permite homogeneizar estas diferencias, asegurando que ninguna variable con valores mayores domine el proceso de agrupamiento.
- Distribuciones controladas:  
Las distribuciones de las variables son en su mayoría sesgadas hacia la izquierda, pero sin grandes anomalías. Esto hace que StandardScaler sea adecuado, ya que la transformación preserva la estructura de los datos.

El DataFrame presenta un desbalance significativo, particularmente en las columnas de Ahorro e Inversión y Financiación, donde las instancias de la clase positiva (clase 1) representan un porcentaje muy reducido del total. Este desbalance puede causar que los modelos predictivos tiendan a favorecer la clase mayoritaria (clase 0), lo que disminuye la capacidad del modelo para identificar correctamente la clase minoritaria, afectando negativamente métricas como el recall y la precisión para la clase positiva.

Para abordar este desequilibrio y mejorar la representación de la clase minoritaria, se evaluarán cinco técnicas de balanceo de datos, cada una con un enfoque diferente:

1. Submuestreo Aleatorio (Undersampling):  
   Esta técnica reduce el número de muestras de la clase mayoritaria seleccionándolas aleatoriamente hasta que ambas clases estén equilibradas. Al trabajar con un conjunto de datos más balanceado, se evita que el modelo se incline demasiado hacia la clase mayoritaria. Sin embargo, esta técnica puede perder información ya que las instancias eliminadas pueden contener patrones importantes para la predicción.

2. Sobremuestreo (SMOTE):  
   SMOTE genera nuevas muestras sintéticas de la clase minoritaria creando combinaciones lineales entre los ejemplos existentes más cercanos en el espacio de características. Esto aumenta la representación de la clase minoritaria sin duplicar los datos reales, lo que reduce el riesgo de sobreajuste que ocurre al simplemente replicar muestras. SMOTE es particularmente útil cuando el DataFrame es pequeño o cuando se desea preservar toda la información disponible.

3. Combinación de Submuestreo y Sobremuestreo (Muestreo Mixto):  
   En algunos casos, combinar ambas técnicas puede ofrecer los mejores resultados. Este enfoque primero utiliza SMOTE para generar nuevas muestras sintéticas de la clase minoritaria, asegurando una representación más equilibrada. Posteriormente, aplica submuestreo aleatorio a la clase mayoritaria, reduciendo su tamaño para evitar que domine el DataFrame. Esta combinación permite mantener un DataFrame más manejable, equilibrado y con menor riesgo de sobreajuste. Además, se mitigan las limitaciones individuales de cada técnica, ya que SMOTE preserva patrones relevantes de la clase minoritaria, mientras que el submuestreo controla el tamaño del conjunto de datos.

4. SMOTEENN:  
   SMOTEENN combina SMOTE con ENN (Edited Nearest Neighbors), una técnica que elimina muestras ruidosas o inconsistentes de ambas clases después del sobremuestreo. Esto ayuda a limpiar el DataFrame y reducir el ruido, mejorando la calidad del balance.

5. SMOTETomek:  
   SMOTETomek combina SMOTE con Tomek Links, una técnica que identifica y elimina pares de instancias (una de cada clase) que están muy cerca en el espacio de características, lo que indica que son ambiguas. Esto mejora la separación entre clases, ayudando al modelo a aprender límites más claros.

Cada una de estas técnicas será evaluada para determinar cuál mejora más el rendimiento del modelo en términos de métricas clave, como accuracy, precision, recall, f1_score y roc_auc, con un enfoque especial en el rendimiento para la clase minoritaria.

### 8.8 Cálculo del ROI
[Volver al índice](#index)

**Cálculo del ROI** se hace según la siguiente fórmula:
$$
ROI = \frac{\text{Ingresos estimados} - \text{Costes totales}}{\text{Costes totales}} \times 100
$$
$$
\text{Ingresos} = \text{Emails enviados}  \times \text{(1-Tasa de rebote)}\times \text{Tasa de apertura}\times \text{Ganancia promedia por compra} 
$$

$$
\text{Ingresos} = \text{Ingresos}_{\text{ahorro e inversión}} + \text{Ingresos}_{\text{financiación}}
$$
$$
\text{Costes totales} = \text{Emails enviados} \times \text{Coste por email}
$$

**Análisis de los supuestos**:

1. **Porcentaje de rebote:**  
   En 2019, el índice de rebote promedio en correos electrónicos para campañas de marketing oscilaba alrededor del 0.7%, dependiendo del ámbito y la calidad de las bases de datos.

2. **Porcentaje de apertura:**  
   En 2019, el índice de apertura promedio de correos electrónicos en campañas de marketing fluctuaba entre el 15% y el 30%, condicionado por el sector y la calidad de las listas.

3. **Beneficio medio por transacción:**  
   Representa el ingreso promedio obtenido por cada operación, calculado para las categorías de ahorro, inversión y financiación.

4. **Precio por correo:**  
   En 2019, los precios de envío de correos electrónicos variaban entre 0.0009 € y 0.01 € por mensaje.

Consideraremos 3 escenarios: base, optimista y pesimista, para analizar cómo pueden variar los resultados según diferentes valores de las variables clave, tales como las tasas de apertura, las tasas de conversión o los costes.

```python
# Ganancia promedio por categoría
ganancia_promedio_ahorro = predicciones_10000[predicciones_10000['producto_optimo'] == 'Ahorro e inversión']['ganancia_maxima'].mean()
ganancia_promedio_financiacion = predicciones_10000[predicciones_10000['producto_optimo'] == 'Financiación']['ganancia_maxima'].mean()

# Datos base
emails_enviados = 10000
emails_ahorro=predicciones_10000[predicciones_10000['producto_optimo'] == 'Ahorro e inversión']['Customer_ID'].count()

# Escenarios
escenarios = {
    "pesimista": {"tasa_apertura": 0.15,"tasa_rebote": 0.015,"coste_por_email": 0.01},
    "base": {"tasa_apertura": 0.225,"tasa_rebote": 0.007,"coste_por_email": 0.001},
    "optimista": {"tasa_apertura": 0.30,"tasa_rebote": 0.003, "coste_por_email": 0.0009}

}

# Lista para almacenar los resultados de cada escenario
resultados = []

# Cálculo por escenario
for escenario, valores in escenarios.items():
    tasa_rebote = valores["tasa_rebote"]
    tasa_apertura = valores["tasa_apertura"]
    coste_por_email = valores["coste_por_email"]
    
    ingresos_ahorro = round(emails_ahorro *(1-tasa_rebote)* tasa_apertura * ganancia_promedio_ahorro, 0).astype(int)
    ingresos_financiacion = round((emails_enviados-emails_ahorro) * (1-tasa_rebote) * tasa_apertura * ganancia_promedio_financiacion, 0).astype(int)
    ingresos_totales = ingresos_ahorro + ingresos_financiacion
    
    costes_totales = emails_enviados * coste_por_email
    roi = round(((ingresos_totales - costes_totales) / costes_totales) * 100,0).astype(int)
    
    # Guardar resultados en una lista
    resultados.append({
        "Escenario": escenario.capitalize(),
        "Tasa rebote": tasa_rebote,
        "Tasa apertura": tasa_apertura,
        "Coste por email (€)": coste_por_email,
        "Ingresos Ahorro (€)": ingresos_ahorro,
        "Ingresos Financiación (€)": ingresos_financiacion,
        "Ingresos Totales (€)": ingresos_totales,
        "Costes Totales (€)": costes_totales,
        "ROI (%)": roi
    })

# Crear un DataFrame con los resultados
df_roi = pd.DataFrame(resultados)

# Mostrar el DataFrame
df_roi
```

Tabla 6: KPIs de interacción 

En el siguiente punto se presenta una lista de KPIs vinculados a los beneficios económicos. Aunque son ficticios, están fundamentados en las estimaciones previas. Uno de los indicadores más destacados es el ROI por cluster, cuyo cálculo se expone a continuación.

```python
# Agrupar los datos por cluster
ganancias_cluster = df_Tarea4_fin.groupby(['cluster_name']).agg(
    # Calcular la media de 'ganancia_maxima' para cada cluster y asignarla a la columna 'Ganancias_cluster'
    Ganancias_cluster=('ganancia_maxima', np.mean),

    # Contar el número de clientes (longitud de la serie 'ganancia_maxima') para cada cluster y asignarlo a 'Num_clientes_cluster'
    Num_clientes_cluster=('ganancia_maxima', lambda series: len(series)),
)
# Mostrar el DataFrame con las ganancias promedio y el número de clientes por cluster
ganancias_cluster
```

```python
# Seleccionar las primeras tres columnas del DataFrame df_roi y asignarlas a presupuestos_roi
presupuestos_roi = df_roi.iloc[:, :4]
# Mostrar el contenido del DataFrame presupuestos_roi para verificar la selección
presupuestos_roi
```

```python
# Iterar sobre los escenarios en el DataFrame presupuestos_roi
for j in range(presupuestos_roi.shape[0]):
    # Iterar sobre los clusters en el DataFrame ganancias_cluster
    for i in range(ganancias_cluster.shape[0]):
        # Obtener el sufijo correspondiente al escenario actual 
        suffix_col = presupuestos_roi.iloc[j, 0]

        # Calcular los ingresos para el cluster en el escenario actual
        # Ingresos = Ganancias del cluster * (1- Tasa de rebote) * Tasa de apertura * Número de clientes del cluster
        ganancias_cluster["Ingresos_cluster_" + suffix_col] = (
            ganancias_cluster['Ganancias_cluster'] * 
            (1-presupuestos_roi.loc[j, 'Tasa rebote']) * 
            presupuestos_roi.loc[j, 'Tasa apertura']*
            ganancias_cluster['Num_clientes_cluster']
        )

        # Calcular los costes para el cluster en el escenario actual
        # Costes = Coste por email del escenario * Número de clientes del cluster
        ganancias_cluster["Costes_cluster_" + suffix_col] = (
            presupuestos_roi.loc[j, 'Coste por email (€)'] * 
            ganancias_cluster['Num_clientes_cluster']
        )

        # Calcular el ROI (Retorno de Inversión) para el cluster en el escenario actual
        # ROI = ((Ingresos - Costes) / Costes) * 100
        ganancias_cluster["ROI_cluster_" + suffix_col] = (
            (ganancias_cluster["Ingresos_cluster_" + suffix_col] - 
             ganancias_cluster["Costes_cluster_" + suffix_col]) / 
            ganancias_cluster["Costes_cluster_" + suffix_col]
        ) * 100
```

```python
# Visualizar los resultados
ganancias_cluster.round(2)
```
